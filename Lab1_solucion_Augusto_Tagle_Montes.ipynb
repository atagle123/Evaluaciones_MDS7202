Responda las siguientes preguntas (0.2 pts c/u):

1. ¿Qué es `Git` y cómo se relaciona con el control de versiones?  
Git es un controlador de versiones (VCS) que permite controlar los cambios que se van haciendo a archivos y permite trabajar colaborativamente sobre esos archivos. 

2. Nombre y explique 3 diferencias entre `Git` y `Github`.  
-Github es un servicio de almacenamiento en la nube, y git es un sistema de control de versiones distribuido.
-Da una interfaz gráfica para los repositorios almacenados, mientras que git a priori es una herramienta de linea de comandos
-Git crea un repositorio local para ver los cambios locales sobre el repositorio, mientras que en github se almacenan copias de repositorios git.

3. ¿Qué es un `hash` y cómo se representa en un `commit`? De un ejemplo.  
Un hash es una cadena de letras y numeros que se genera con una función de hash, en donde la funcion toma ciertos parametros de entrada y devuelve siempre una cadena del mismo largo, y ante el menor cambio en los parametros de entrada devuelve una cadena totalmente distinta, por lo que sirve como identificador unico. En un commit, el hash (sha-1) sirve para identificar únicamente a ese commit. Por ejemplo: 
commit 1 -->  10a5b8aab34f7f8d475cfa3e551d3j5y78g04b3de5
4. ¿Que elementos componen a la estructura de datos de `Git`?  
Lo componen los blobs, que son los archivos en cada punto del tiempo, y se identifcan con un hash.  
Tambien lo componen los arboles, que componen arboles de blobs para ver las ramas de las versiones que hay de los archivos.


5. ¿Que comandos de `Git` sirven para crear un repositorio y hacer seguimiento a cambios en los archivos?  
Para crear un repositorio se puede utilizar git init o git clone si se quiere clonar un repositorio de algun servicio de almacenamiento como github.
Para hacer seguimiento a los cambios de los archivos se puede utilizar git status, que entrega los archivos que han sido modificados. Tambien se puede utilizar git log que muestra el historial de commits. Y por último tenemos git diff que muestra las diferencias entre los archivos del repositorio.

11. `git stash` (0.25 puntos)

Git stash sirve para guardar los cambios de archivos, temporalmente y sin un commit. Por lo que permite almacenar cambios sin tener que hacer un commit y posteriormente trabajar sobre ellos.

12. `git reset`. ¿Cuál es la diferencia de este comando con `git revert`?. (0.25 puntos)

Git reset sirve para volver hacia "atras" en el historial de commits, lo que sirve para deshacer cambios en los arhchivos. Se diferencia con git revert, ya que git revert deshace los cambios pero creando un nuevo commit. Por lo que seria posible volver al commit inicial. Con git reset esto es mas dificil.
