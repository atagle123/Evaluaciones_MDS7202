```python
import pandas as pd
import os
import shutil
import datetime
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder,FunctionTransformer, OrdinalEncoder 
from sklearn.pipeline import Pipeline
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as ss
from scipy.stats import chi2_contingency
%matplotlib inline
```

```python
class Profiler:
    def __init__(self,df):
        """
        Clase que se le entrega un dataframe, y crea una carpeta donde se guardaran distintos analisis de los datos
        """
        self.df=df
        nombre_carpeta="EDA_"+datetime.datetime.now().strftime("%d-%m-%Y")  #nombre carpeta
        ruta_carpeta = os.path.join(os.getcwd(),nombre_carpeta)
        self._ruta_carpeta=ruta_carpeta
        self.__create_folder_route()  # llama a funci√≥n que crea carpeta
        pass

    def __create_folder_route(self,nombre_carpeta=None):
        """
        Funci√≥n que crea la carpeta en el directorio, si no hay nombre de carpeta, crea el directorio
        """
        if nombre_carpeta==None:
            ruta_carpeta=self._ruta_carpeta
            nombre_carpeta=ruta_carpeta
        else:
            ruta_carpeta=f"{self._ruta_carpeta}/{nombre_carpeta}"
        
        if not os.path.exists(ruta_carpeta):
            os.mkdir(ruta_carpeta)
            print(f'Se ha creado la carpeta "{nombre_carpeta}" en el directorio actual.')
        else:
            print(f'La carpeta "{nombre_carpeta}" ya existe en el directorio actual.')
        pass
    def __select_variables(self,variables):
        """ 
        Selecciona columnas del dataframe, de una lista de variables.
        """
        try:
            df=self.df[variables]
        except:
            df=self.df
        return(df)
    
    def __drop_duplicated(self,df):
        """ 
        Elimina valores duplicados en dataframe
        """
        return(df.drop_duplicates())

    def __separate_non_atomic(self,df):
        """
        Separa las columnas no √°tomicas de un dataframe
        """
        columnas_no_atomicas = df.filter(regex='-', axis=1).columns.values # deje solo "-" como separador
        for i in columnas_no_atomicas:
            df[i.split("-")] = df[i].str.split(r"\*|\?|:|\(", expand=True)  #expresion regular r"\*|\?|:|\(" para separar las columnas
            df = df.drop(columns=[i]) 
        df = df.replace('nan', np.nan)  
        df = df.apply(pd.to_numeric, errors='ignore')
        return(df)
    
    def __fillnan(self,df,method="None"): # acepta 3 metodos, se pueden agregar mas.
        """
        Llena los valores Nan de un dataframe con el m√©todo seleccionado
        """
        if method=="mode":   # cambiar valores nulos por la moda
            df = df.apply(lambda col: col.fillna(col.mode()[0]), axis=0)
            
        elif method=="None":  # no hacer nada con los valores nulos para las variables categoricas y cambiar por moda en variables numericas
            cat_columns = df.select_dtypes(include=['object']).columns
            df[cat_columns] = df[cat_columns].fillna("None")
            
            num_columns = df.select_dtypes(include=['number']).columns
            df[num_columns] = df[num_columns].apply(lambda col: col.fillna(col.mode()[0]), axis=0)

        else: # eliminar valores nulos
            df=df.dropna()   
        return(df)
    
    def __scale_data(self,df,cat_scaler):
        """
        Escala los valores n√∫mericos y categoricos
        """
        numeric_columns = df.select_dtypes(include=['number']).columns
        categorical_columns = df.select_dtypes(include=['object']).columns

        numeric_preprocessing = Pipeline([
            ('log', FunctionTransformer(np.log1p)),  # log 
            ('min-maxscaler', MinMaxScaler())   # min max scaler
        ])

        ct = ColumnTransformer(transformers=[
            ("num", numeric_preprocessing, numeric_columns),
            ("scaler_cat", cat_scaler, categorical_columns )]
        )
        transformed_data = ct.fit_transform(df)
        df=pd.DataFrame(transformed_data, columns=df.columns)
        return(df)
    
    def __optimal_clusters(self,df, max_clusters=10): 
        """
        Funci√≥n que encuentra el n√∫mero √≥ptimo de clusters, utilizando el m√©todo del codo.
        """
        distortions = [KMeans(n_clusters=i, random_state=0).fit(df).inertia_ for i in range(1, max_clusters + 1)]
        plt.plot(range(1, max_clusters + 1), distortions, marker='o')
        plt.title('M√©todo del Codo')
        plt.xlabel('N√∫mero de clusters')
        plt.ylabel('Distorsi√≥n')
        plt.grid()
        plt.show()
        n_clusters=int(input())
        return(n_clusters)
    def __make_plot_clusters(self,df,clusters):
        """
        Crea un plot de clusters en 2d utilizando PCA
        """
        pca = PCA(n_components=2, random_state=123)
        pca_2d = pca.fit_transform(df)
        plt.figure(figsize=(8, 6))
        plt.scatter(pca_2d[:, 0], pca_2d[:, 1], c=clusters, cmap='viridis')
        plt.title('PCA Visualization of Clusters')
        plt.xlabel("PC1")
        plt.ylabel("PC2")

        plt.colorbar()
        return(plt)
    def __isolation_forest(self,df):
        """
        A partir de un dataframe busca anomalias con isolation forest, devuelve anomalias.
        """
        columnas_numericas = df.select_dtypes(include=['number'])
        scaler = StandardScaler() 
        scaled_data = scaler.fit_transform(columnas_numericas) #esacala los datos antes del modelo

        model = IsolationForest()  # Entrena un Isolation Forest para detectar anomal√≠as

        model.fit(scaled_data)
        anomaly = model.predict(scaled_data) # Predice anomal√≠as en los datos

        return(anomaly)
    def __v_de_cramer(self,confusion_matrix):
        """ 
        Funcion para calcular coeficiente v de cramer, obtenido de:
            Journal of the Korean Statistical Society 42 (2013): 323-328
        """
        chi2 = chi2_contingency(confusion_matrix)[0]
        n = confusion_matrix.sum().sum()
        phi2 = chi2 / n
        r, k = confusion_matrix.shape
        phi2corr = max(0, phi2 - ((k-1)*(r-1)) / (n-1))
        rcorr = r - ((r-1)**2) / (n-1)
        kcorr = k - ((k-1)**2) / (n-1)
        return np.sqrt(phi2corr / (min((kcorr-1), (rcorr-1))))

    def __cramers_matrix(self,df):
        """
        Genera matriz con coeficientes v de cramer
        """
        n = df.shape[1]
        matrix = np.zeros((n, n))
        for i in range(n-1):
            for j in range(n-1):
                if i == j:
                    matrix[i, j] = 1.0
                else:
                    confusion_matrix = pd.crosstab(df.iloc[:, i], df.iloc[:, j])
                    matrix[i, j] = self.__v_de_cramer(confusion_matrix)
        return matrix

    def summarize(self,variables=None,anomaly_treshold=0,clean=False,method="None"):
        """
        M√©todo que resume la lista de variables entregadas
        """
        if clean:   # elejir si hacer summary de variables limpias o no
            df_variables=self.clean_data(variables=variables,method=method)
        else:
            df_variables=self.__select_variables(variables) #selecciona variables
        df_info=pd.DataFrame({ "non-nulls": len(df_variables)-df_variables.isnull().sum().values, "nulls": df_variables.isnull().sum().values, "type": df_variables.dtypes.values,"unique_values":df_variables.nunique()}) # se calculan algunos datos del dataframae

        columnas_numericas = df_variables.select_dtypes(include=['number'])  # selecciona columnas numericas
        df_numeric=columnas_numericas.describe().T
        df_numeric = df_numeric.assign(Zeros=(columnas_numericas == 0).sum(), Negatives=(columnas_numericas < 0).sum(),Outliers=columnas_numericas.apply(lambda col: ((abs(col-col.mean()) > 3* col.std()).sum())))  # calcula estadisticas de variables numericas

        merged_df = pd.merge(df_info, df_numeric, left_index=True, right_index=True, how='outer')

        anomalies=self.__isolation_forest(df_variables) # usaremos isolation forest para detectar anomalias
        diff_anomalies=np.sum(anomalies)
        if diff_anomalies<anomaly_treshold:  # -1 son anomalias y 1 no son anomalias, si hay mas anomalias que no anomalias, levantara una alerta
            print(f"ALERTA DE ANOMALIAS! Hay {-diff_anomalies} mas anomalias que no anomalias ") ##ver si cambiar esto y revisar que funque
        merged_df.to_csv(f"{self._ruta_carpeta}/summary.txt", sep='\t', index=True)
        pass
    
    def plot_vars(self,variables=None,n=10,clean=False,method="None"): 
        """
        Genera imagenes con gr√°ficos de interes, ditribucion de variables, correlacion de variables, histograma con top N categorias y matriz v de cramer.

        """
        if clean:   # elejir si hacer plot de variables limpias o no
            df_variables=self.clean_data(variables=variables,method=method)
        else:
            df_variables=self.__select_variables(variables) #selecciona variables

        self.__create_folder_route(nombre_carpeta="plots") # crea carpeta plots

        columnas_numericas = df_variables.select_dtypes(include=['number'])
        for columna in columnas_numericas.columns.values:
            ax=sns.displot(df_variables[columna], kde=True, color='blue').set(title=f'Distribuci√≥n de {columna}')
            plt.savefig(f"{self._ruta_carpeta}/plots/{columna}.pdf",format="pdf",bbox_inches='tight')
            plt.clf()

        correlation_matrix = columnas_numericas.corr()
        ax=sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5).set(title=f'Matriz de correlaci√≥n')
        plt.savefig(f"{self._ruta_carpeta}/plots/Correlation_matrix.pdf",format="pdf",bbox_inches='tight')
        plt.clf()

        categorical_columns = df_variables.select_dtypes(include=['object'])
        for columna in categorical_columns.columns.values:
            top_categories = df_variables[columna].value_counts().nlargest(n)
            plt.bar(top_categories.index, top_categories.values)
            plt.title(f"Histograma top {n} categorias")  
            plt.xticks(rotation = 45)
            plt.savefig(f"{self._ruta_carpeta}/plots/{columna}.pdf",format="pdf",bbox_inches='tight')
            plt.clf()

        cramer_matrix=self.__cramers_matrix(categorical_columns)
        cramer_df = pd.DataFrame(cramer_matrix, columns=categorical_columns.columns, index=categorical_columns.columns)
        ax=sns.heatmap(cramer_df, annot=True, cmap='coolwarm', linewidths=.5)
        plt.savefig(f"{self._ruta_carpeta}/plots/v_cramer.pdf",format="pdf",bbox_inches='tight')
        plt.clf()
        pass
    def clean_data(self,variables=None,method="None"):
        """
        M√©todo para limpiar los datos 
        """
        self.__create_folder_route(nombre_carpeta="clean_data")

        df_variables=self.__select_variables(variables)

        pipeline = Pipeline([
                        ('drop_duplicated', FunctionTransformer(self.__drop_duplicated)),
                        ('separate_non_atomic', FunctionTransformer(self.__separate_non_atomic)),
                        ('fillna', FunctionTransformer(self.__fillnan,kw_args={'method': method}))
                    ])
        df_clean=pipeline.fit_transform(df_variables)
        df_clean.to_csv(f"{self._ruta_carpeta}/clean_data/data.csv", index=True)
        return(df_clean)
    
    def scale(self,variables=None,cat_scaler=OrdinalEncoder(dtype=np.int64),method="None"):
        """
        M√©todo para escalar las variables categoricas y n√∫mericas
        """
        self.__create_folder_route(nombre_carpeta="scale")

        df_clean=self.clean_data(variables=variables,method=method)

        df_scaled=self.__scale_data(df_clean,cat_scaler=cat_scaler)
        df_scaled.to_csv(f"{self._ruta_carpeta}/scale/scaled_features.csv", index=True)
        return(df_scaled)
    
    def make_clusters(self,clustering_algorithm,variables=None,method="None"):
        """
        M√©todo para hacer clusters de variables, llama al m√©todo optimal clusters, que encuentra el n√∫mero √≥ptimo de clusters con el m√©todo del codo
        """

        self.__create_folder_route(nombre_carpeta="clusters")
        df_variables=self.__select_variables(variables)

        pipeline = Pipeline([
                ('drop_duplicated', FunctionTransformer(self.__drop_duplicated)),
                ('separate_non_atomic', FunctionTransformer(self.__separate_non_atomic)),
                ('fillna', FunctionTransformer(self.__fillnan,kw_args={'method': method})),
                ('scaler', FunctionTransformer(self.__scale_data,kw_args={'cat_scaler': OrdinalEncoder(dtype=np.int64)}))
            ])

        df_preprocesed = pipeline.fit_transform(df_variables)
        n_clusters=self.__optimal_clusters(df_preprocesed) # comprobar que funcione bien
        print(f"n_clusters:{n_clusters}")
        clustering_algorithm.set_params(n_clusters=n_clusters)  
        clusters=clustering_algorithm.fit(df_preprocesed)
        plt_clusters=self.__make_plot_clusters(df_preprocesed,clusters.labels_)
        df_preprocesed["clusters"]=clusters.labels_ 
        plt_clusters.savefig(f"{self._ruta_carpeta}/clusters/data_clusters.png", format='png')
        plt_clusters.close()
        df_preprocesed.to_csv(f"{self._ruta_carpeta}/clusters/data_clusters.csv", index=True)
        pass
    
    def detect_anomalies(self,anomalies_algorithm=IsolationForest(contamination=0.05),variables=None,method="None"):
        """
        M√©todo para detectar √°nomalias en los datos
        """
        self.__create_folder_route(nombre_carpeta="anomalies")
        df_summary=self.__select_variables(variables)

        pipeline = Pipeline([
                ('drop_duplicated', FunctionTransformer(self.__drop_duplicated)),
                ('separate_non_atomic', FunctionTransformer(self.__separate_non_atomic)),
                ('fillna', FunctionTransformer(self.__fillnan,kw_args={'method': method})),
                ('scaler', FunctionTransformer(self.__scale_data,kw_args={'cat_scaler': OrdinalEncoder(dtype=np.int64)}))
            ]) 
        df_preprocesed = pipeline.fit_transform(df_summary)

        columnas_numericas = df_preprocesed.select_dtypes(include=['number'])
        anomalies_algorithm.fit(columnas_numericas)
        anomalies = anomalies_algorithm.predict(columnas_numericas)
        
        plt_clusters=self.__make_plot_clusters(df_preprocesed,anomalies)
        df_preprocesed["anomalies"]=anomalies
        plt_clusters.savefig(f"{self._ruta_carpeta}/anomalies/data_anomalies.png", format='png')
        plt_clusters.close()
        df_preprocesed.to_csv(f"{self._ruta_carpeta}/anomalies/data_anomalies.csv", index=True)
        pass

    def profile(self,variables=None,clean=False,method="None",clustering_algorithm=KMeans(n_clusters=2, random_state=1)):
        """
        Ejecuta todos los metodos anteriores correspondientes al analisis de datos
        """
        self.summarize(variables=variables,clean=clean)
        self.plot_vars(variables=variables,clean=clean,method=method)
        self.clean_data(variables=variables,method=method)
        self.scale(variables,method=method)
        self.make_clusters(clustering_algorithm=clustering_algorithm,variables=variables,method=method)
        self.detect_anomalies(variables=variables,method=method)
        pass

    def clearGarbage(self):
        """
        Funcion para eliminar todos los archivos creados por esta clase.
        """
        for root, dirs, files in os.walk(os.getcwd()):
            for directorio in dirs:
                if directorio.startswith("EDA"):
                    ruta_completa = os.path.join(root, directorio)
                    try:
                        shutil.rmtree(directorio)
                        print(f"{directorio} eliminado correctamente")
                    except Exception as e:
                        print(f"No se pudo eliminar el directorio {ruta_completa}: {e}")
```
